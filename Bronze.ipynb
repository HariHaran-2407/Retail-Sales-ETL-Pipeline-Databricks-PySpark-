### **Incremental Load Initial Step**

```python
#incremental load query
if spark.catalog.tableExists('retailsales.bronze.bronze_table'):
  last_load_date= spark.sql(""" SELECT MAX(last_updated) FROM retailsales.bronze.bronze_table""").collect()[0][0]
else:
  last_load_date='1100-01-01';
```

```python
last_load_date
```

```python
%sql
SELECT * FROM retailsales.default.source_data
```

```python
#temporary view for incremented load rows 
spark.sql(f"""SELECT * FROM retailsales.default.source_data
where last_updated> '{last_load_date}'""").createOrReplaceTempView("bronze_source")
```

```python
%sql
select * from bronze_source
```

```python
%sql
--creating actual bronze_table
create or replace table retailsales.bronze.bronze_table
as
select * from bronze_source;
```
